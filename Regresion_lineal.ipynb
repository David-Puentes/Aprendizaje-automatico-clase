```python
location = r"student.csv"
data=pd.read_csv(location,sep=";")
data[:5]
Math= data["Math"].values
Reading= data["Reading"].values
Writing= data["Writing"].values
```

```python
m=len(Math)
x0=np.ones(m)
X=np.array([x0,Math,Reading]).T

# Coeficientes iniciales
B=np.array([0,0,0])
Y=np.array(Writing)
alpha= 0.0001
Y.shape
```

```python
# Funcin de costo
def cost_func(X,Y,B):
    m=len(Y)
    J=np.sum((X.dot(B)-Y)**2)/(2*m)
    return J
```

```python
# Definicin del gradiente

def grad(X,Y,B,alpha,iterations):
    cost_hist=[0]*iterations
    m=len(Y)
    for i in range(iterations):
        h=X.dot(B)
        loss= h-Y
        gradient= X.T.dot(loss)/m
        B=B-alpha*gradient
        cost=cost_func(X,Y,B)
        cost_hist[i]=cost
    return B, cost_hist
```

```python
newB, cost_hist=grad(X,Y,B,alpha,1000000)
print(newB)
print(cost_hist[-1])
```
